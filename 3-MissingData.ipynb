1 --> Finding Missing Data.    .isna()
2 --> Removing Missing Data.   .dropna()
3 --> Filling Missing Data.    .fillna()

we need to use inplace=True to make the changes permanent

```python
import numpy as np 
import pandas as pd 
```

```python
# Finding missing Data

data = {
    'A' : [1,2,np.nan,4,5],
    'B' : [1,2,3,4,5],
    'C' : [1,2,3,np.nan,np.nan],
    'D' : [1,np.nan,np.nan,np.nan,5]
}

df = pd.DataFrame(data)
df
```

```python
# Check is we have any null or nan values in dataframes
# use .isna()

df.isna()
```

```python
# Check how many null values we have in each column
df.isna().sum()
```

### Removing missing data
It only works in rows basis

```python
df
```

```python
df.dropna()    #This will remove all the rows which contains null values
```

```python
df
```

```python
df.dropna(thresh=3) #Thresh states that we need atleast 3 non-null values in the row
```

### Filling the missing data

```python
df
```

```python
df.fillna(0)    #Fill every null value with 0
```

```python
val = {
    'A' : 100,
    'B' : 200,
    'C' : 300,
    'D' : 400
}

df.fillna(value=val)  #Fills the null value accordung to dict we provide 
```

```python
df.fillna(df.mean())   #Fill the null values with the mean 
```
