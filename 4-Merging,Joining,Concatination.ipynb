```python
import numpy as np
import pandas as pd
```

```python
employees = {
    'Id' : [1,2,3,4,5],
    'Name' : ['John','Ana','Peter','linda','bob'],
    'Dept' : ['HR','IT','FINANCE','IT','HR']
}

salaries = {
    'Id' : [1,2,3,6,7],
    'sal' : [60000,80000,65000,70000,90000],
    'Bonus' : [5000,10000,7000,8000,12000]
}

empDF = pd.DataFrame(employees)
salDF = pd.DataFrame(salaries)
```

```python
empDF
```

```python
salDF
```

### Merging

```python
pd.merge(empDF,salDF,on = 'Id',how = 'inner')   #inner only gives the values which have all the data. & on = 'Id' defines on which basis merge should happen
```

```python
pd.merge(empDF,salDF,on = 'Id' , how = 'outer')   #How = 'outer' will keep missing values null & on='Id' defines on which basis or common column merge should happen
```

### Concatination

```python
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2'],
    'B' : ['B0','B1','B2'],
    'C' : ['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A' : ['A3','A4','A5'],
    'B' : ['B3','B4','B5'],
    'C' : ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
# Concatinate 2 df vertically

pd.concat([df1,df2])
```

```python
# Concatinate 2 df horizontally

pd.concat([df1,df2],axis = 1)
```
