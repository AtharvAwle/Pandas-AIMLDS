### Creating Series

```python
import pandas as pd
import numpy as np
```

```python
label = ['a','b','c']
l1 = [10,20,30]
arr = np.array([40,50,60])
dic = {'d': 70 , 'e' : 80 , 'f' : 90}
```

```python
# Series with list without custom index label
pd.Series(l1)
```

```python
# series with list with custom label index
pd.Series(l1,index = label)
```

```python
# Series with array
pd.Series(arr , index = label)
```

```python
# Series wirh dictionary

pd.Series(dic)
```
