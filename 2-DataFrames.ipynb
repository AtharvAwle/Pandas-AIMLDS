### DataFrames In Pandas
A Pandas DataFrame is a two-dimensional table-like structure in Python where data is arranged in rows and columns. Itâ€™s one of the most commonly used tools for handling data and makes it easy to organize, analyze and manipulate data.

It is Collection of multiple series

```python
import numpy as np 
import pandas as pd
```

#### Creating DataFrames

```python
# Creating a DataFrame using Dictionary

data = {
    'Name' : ['Atharv' , 'Ayush' , 'Shlok' , 'Atul' , 'Srushti'],
    'Age' : [21 , 15 , 22 , 23 , 24],
    'City' : ['Palava' , 'Thane' , 'Deslepada' , 'Lodha' , 'Sagaon'],
    'Salary' : [10000 , 20000 , 30000 , 40000 , 50000]
}

dictDF = pd.DataFrame(data)
dictDF
```

```python
# Creating a DataFrame using List

l1 = [
    ['Atharv' , 21 , 'Palava' , 10000],
    ['Ayush' , 15 , 'Thane' , 20000],
    ['Shlok' , 22 , 'Deslepada' , 30000],
    ['Atul' , 23 , 'Lodha' , 40000],
    ['Srushti' , 24 , 'Sagaon' , 50000] 
]

col = ['Name' , 'Age' , 'City' , 'Salary']   #In list as we dont have any keys we have to add colums

listDF = pd.DataFrame(l1,columns=col)
listDF
```

### Selection and Indexing of Columns

```python
dictDF['Name'] #Target Name column from the dataframe
dictDF[['Age' , 'City']] #Target Name & City from dataframe 
```

### Creating New Column

```python
# Adding one more column in our dataframe
dictDF['Designation'] = ['Engg' , 'Doctor' , 'MD' , 'CA' , 'CA']
```

```python
dictDF
```

### Removing Column

```python
dictDF
```

```python
# Removing designation from the DF

dictDF.drop('Designation' , axis=1 , inplace=True)  #by using inplace it will permanently delete the column
```

```python
dictDF
```
